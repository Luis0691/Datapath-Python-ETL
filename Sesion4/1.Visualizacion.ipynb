```python
import matplotlib.pyplot as plt
```

```python
# Datos
x = [1, 2, 3, 4, 5]
y = [3, 4, 6, 7, 8]

plt.plot(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title("Grafico de linea")

plt.show()
```

```python
# Grafico de dispersion

plt.scatter(x, y)
plt.title("Grafico de Dispersion")
plt.show()
plt.savefig("grafico_dispersion.jpg")
```

```python
import numpy as np

datos = np.random.randn(1000)

plt.hist(datos, color='skyblue', edgecolor='black', bins=30 )
plt.title('Histograma')
plt.show()
```

```python
# Grafico de barras

categorias = ['A', 'B', 'C', 'D']
valores = [24, 56, 23, 90]

plt.bar(categorias, valores, color='green')
plt.show()
```
